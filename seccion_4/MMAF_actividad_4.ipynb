#### Actividad: Explorando la vida útil del cable

Imagina que en Buenaventura se utiliza un cable submarino para transmitir datos. El cable tiene dos aspectos importantes:

1. **Temperatura del cable:**  
   La temperatura depende del flujo de datos $D$ (en unidades arbitrarias) según la función:
   $$
   T(D) = 8D + B,
   $$
   donde $B$ es una constante que se puede ajustar. El cable tiene una **temperatura crítica** de $T_c = 50^\circ$ grados, temperatura en la cual el cable empieza a derretirse y por lo tanto debe evitarse.

2. **Capacidad máxima de transmisión**  
   El envejecimiento del cable submarino es un problema importante en la transmisión de datos, ya que la capacidad de transmisión disminuye con el tiempo. La capacidad de transmisión del cable se puede modelar como una función lineal en función del tiempo $t$ en años:

  $$
   D(t) = -\frac{9}{10}t + 10,
$$
   donde $t$ es el tiempo en años  y la función $D(t)$ representa la capacidad máxima de transmisión de datos en unidades arbitrarias. Para la empresa es viable que la capacidad máxima de transmisión de datos  sea mayor a $D_c=5$, de lo contrario la empresa asumiría perdidas.
